```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
ratio = np.array( [1, 3, 5, 7, 9, 10, 15], dtype = float )

area = np.array([3.14, 28.26, 78.5, 153.86, 254.34, 314, 706.5], dtype = float)
```

```python
input = tf.keras.Input( shape = [1] )
#define explicitly the input to the model

layer = tf.keras.layers.Dense( units=1, activation='relu' )(input)
#add the dense layer connecting it directly with the input

output = tf.keras.layers.Dense(units=1)(layer)

model = tf.keras.Model( inputs = input, outputs = output)
```

```python
model.compile(
    optimizer = tf.keras.optimizers.Adam()
)
```

```python
model.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = 'mean_squared_error'
)
```

```python
print("Training starting")
history = model.fit( ratio, area, epochs= 1500, verbose = False)
print("Model trained")
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(history.history["loss"])
```

```python
print("hagamos prediccion")

predictInput = np.array([1.0])

resp = model.predict(predictInput)

resp
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
np.random.seed(42)
radios = np.random.uniform(0, 10, 1000)  # Genera 1000 radios aleatorios entre 0 y 10
areas = np.pi * (radios ** 2)  # Calcula las áreas de los círculos con esos radios
```

```python
areas
```

```python
model = Sequential([
    Dense(64, input_shape=(1,), activation='relu'),
    Dense(16, activation='relu'),
    Dense(1) 
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
print("Training starting")
history = model.fit(radios, areas, epochs=800, batch_size=32, verbose=1)
print("Model trained")
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(history.history["loss"])
```

```python
test_radios = np.array([2.0])  # Algunos ejemplos de radios de prueba
predicted_areas = model.predict(test_radios)

for r, area in zip(test_radios, predicted_areas):
    print(f"Radio: {r}, Área predicha: {area[0]:.2f}, Área real: {np.pi * r ** 2:.2f}")
```
