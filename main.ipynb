```python
import tensorflow as tf
import numpy as np
```

```python
# print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
celcius = np.array( [-40, -10, 0, 8, 15, 22, 38], dtype = float )

fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype = float)
```

```python
#capa = tf.keras.layers.Dense( units = 1, input_shape = [1])

#modelo = tf.keras.Sequential( [ capa ] )
#there is a warning, this says if we use model sequential is not recommend use input_shape
```

```python
entrada = tf.keras.Input( shape = [1] )
#define explicitly the input to the model

capa = tf.keras.layers.Dense( units=1 )(entrada)
#add the dense layer connecting it directly with the input

modelo = tf.keras.Model( inputs = entrada, outputs = capa)
# Create the model using the functional API of keras, this case allows you to 
# specify the inputs and outputs more clearly
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam()
)
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = 'mean_squared_error'
)
```

```python
print("Training starting")
historial = modelo.fit( celcius, fahrenheit, epochs= 1000, verbose = False)
print("Model trained")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
print("hagamos prediccion")

predictInput = np.array([100.0])

resp = modelo.predict(predictInput)

resp
```
